```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
temp = pd.read_csv('E:/Data_Mining_project/data/temperature_data.csv')
temp.head()
```

```python
# melt temp except country name and indicator name
cols_to_melt = [col for col in temp.columns if col not in ['Country Name', 'Indicator Name']]
temp = pd.melt(temp,
                id_vars=['Country Name','Indicator Name'],
                value_vars=cols_to_melt,
                var_name='Year',
                value_name='Temperature')
temp.head()
```

```python
temp.columns
```

```python
regional_temp = temp.groupby(['Indicator Name', 'Year'])['Temperature'].mean().reset_index()
regional_temp['Year'] = regional_temp['Year'].astype(int)
fig = px.line(regional_temp, x='Year', y='Temperature', color='Indicator Name',
              title='Average Temperature of World in Degrees Celsius')
fig.update_traces(line=dict(width=3), marker=dict(size=8))
fig.update_layout(legend_title_text = 'Region', plot_bgcolor='white')
fig.show()
```

```python
# plot regional plots with plotly line
# top 20 countries with highest average temperature
top_countries = temp.groupby('Country Name')['Temperature'].mean().sort_values(ascending=False).head(20).index
temp_top = temp[temp['Country Name'].isin(top_countries)].copy()
temp_top['Year'] = temp_top['Year'].astype(int)
fig = px.line(temp_top, x='Year', y='Temperature', color='Country Name', title='Average Temperature by Country')
fig.show()
```

```python
# south asia plot with plotly for each country in south asia
countries = ['India', 'Pakistan', 'Bangladesh', 'Srilanka', 'Afghanistan'] 
south_asia = temp[temp['Country Name'].isin(countries)].copy()
south_asia['Year'] = south_asia['Year'].astype(int)
fig = px.line(south_asia, x='Year', y='Temperature', color='Country Name', title='Average Temperature in South Asia')
fig.show()
```
