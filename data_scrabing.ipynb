```python
data_filteration = pd.read_csv('E:/Data_Mining_project/data/collectied_data.csv')
data_filteration.shape
```

```python
data_filteration = data_filteration.loc[(data_filteration['Indicator Name'] == 'Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e)')
                                                         |(data_filteration['Indicator Name'] == 'Carbon dioxide (CO2) emissions excluding LULUCF per capita (t CO2e/capita)')
                                                         |(data_filteration['Indicator Name'] == 'PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)')
                                                         |(data_filteration['Indicator Name'] == 'Total greenhouse gas emissions excluding LULUCF (% change from 1990)')
                                                         |(data_filteration['Indicator Name'] == 'Total greenhouse gas emissions excluding LULUCF per capita (t CO2e/capita)')
                                                         |(data_filteration['Indicator Name'] == 'Life expectancy at birth, total (years)')
                                                         |( data_filteration['Indicator Name'] == 'Carbon dioxide (CO2) emissions from Industrial Combustion (Energy) (Mt CO2e)')]
data_filteration.drop(columns=['Unnamed: 0'], inplace=True)
data_filteration.columns
```

```python
# imputing the missing values with mean value
data_imputed = data_filteration.copy()
data_imputed.iloc[:, 2:] = data_imputed.iloc[:, 2:].apply(lambda x: x.fillna(x.mean()), axis=1)
data_imputed = data_imputed.dropna(subset=data_imputed.columns[2:],how='all')
data_imputed.shape
```
